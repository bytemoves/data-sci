```python
import re

def extract_emails_from_file(file_path):
    with open(file_path, 'r') as file:  # Use the provided file_path parameter
        text = file.read()
        email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'
        emails = re.findall(email_pattern, text)
        return emails

# Provide the path to your text file
file_path = r'D:\pandas\florida.gov.txt'  # Remove the quotes and escape the backslashes using 'r'
emails = extract_emails_from_file(file_path)

# Print the extracted emails
for email in emails:
    print(email)

```

```python
import csv
import re

def extract_emails_from_file(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
        email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'
        emails = re.findall(email_pattern, text)
        return emails

# Provide the path to your text file
file_path = r'D:\pandas\florida.gov.txt'
emails = extract_emails_from_file(file_path)

# Specify the output CSV file path
output_csv_path = r'D:\pandas\mailings.csv'

# Write emails to CSV file
with open(output_csv_path, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Emails'])
    for email in emails:
        writer.writerow([email])

print(f"Emails saved to {output_csv_path}")
```
